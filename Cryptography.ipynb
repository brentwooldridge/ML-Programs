```python
def caesar(x):
    return (x-3)%5
caesar(1)
```

```python
def cinverse(t):
    return (t+3)%5
cinverse(caesar(0))
```

```python
for i in range (0, len(list('abcdefghijklmnopqrstuvwxyz'))):
    print([list('abcdefghijklmnopqrstuvwxyz')[caesar(i)], list('abcdefghijklmnopqrstuvwxyz')[cinverse(caesar(i))]])
   
    
    
```

The following are all ways to create chaotically encrypted files.  Apply the right hash function and you can achieve perfect secrecy.

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

xn = 0.1
yn = 0.1
alpha = 0.9

a = []
b = []
r = 0
while r < 1000:

#    xn = np.linspace(0, 1, 1000)
#    xnn = np.linspace(0.5, 1, 1000)
    yn1 = yn-alpha*np.tan(xn)
    xn1 = np.sin(xn) + yn1

    yn = yn1
    xn = xn1
    a.append(xn)
    b.append(yn)
    
    
    r+=1
plt.plot(a,b)
plt.show()
print(a, b) 

```

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

xn = 0.1
yn = 0.1
alpha = 0.9
b = 0.3
a = 1.4
x = []
y = []
r = 0
while r < 1000:

#    xn = np.linspace(0, 1, 1000)
#    xnn = np.linspace(0.5, 1, 1000)
    yn1 = b**2*xn
    xn1 = xn**2 + yn**2 - a**2

    yn = yn1
    xn = xn1
    x.append(xn)
    y.append(yn)
    
    
    r+=1
plt.plot(x,y)
plt.show()
print(x, y) 

```
